# Changing Colorspaces

There are many  functions available for converting colorspaces in OpenCV
These are cv2.cvtColor() and cv2.inRange() etc.

### Program to convert an image into Grayscale

```python
# Step 1:- Import the required packages
import cv2
```

```python
cv2.startWindowThread()
```

```python
# %matplotlib inline
# from matplotlib import pyplot as plt
print("OpenCV Version : %s " % cv2.__version__)
```

```python
# Step 2:- Load the images
img1=cv2.imread('goruntu01.jpg',1)
```

```python
# Step 3:-  Convert the color into GRAYSCALE
gray=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
```

```python
# Step 4:- Show the converted image to the user
cv2.startWindowThread()
cv2.imshow("Converted Image",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Step 5:- Save the image to the drive
cv2.imwrite("grayedimage01.jpg",gray)
```

### Program to convert an image into Grayscale

```python
# Step 1:- Import the required packages
import cv2
# Step 2:- Load the images
img1=cv2.imread("goruntu01.jpg",1)
# Step 3:-  Convert the color into HSV
hsv=cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
# Step 4:- Show the converted image to the user
cv2.startWindowThread()
cv2.imshow("Converted Image",hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
# Step 5:- Save the image to the drive
cv2.imwrite("HSVimage.jpg",hsv)
```

### Program to convert an image into RGB

```python
# Step 1:- Import the required packages
#import cv2 - already imported at the above
# Step 2:- Load the images
img1=cv2.imread("goruntu01.jpg",1)
# Step 3:-  Convert the color into HSV
rgb=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
# Step 4:- Show the converted image to the user
cv2.startWindowThread()
cv2.imshow("Converted Image",rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
# Step 5:- Save the image to the drive
cv2.imwrite("RGBimage.jpg",rgb)
```
